```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_air = 1.
n_water = 1.334
```

```python
def r_TE(ni, theta_i, nt, theta_t):
    return (ni * np.cos(theta_i) - nt * np.cos(theta_t)) \
        / (ni * np.cos(theta_i) + nt * np.cos(theta_t))

def r_TM(ni, theta_i, nt, theta_t):
    return (nt * np.cos(theta_i) - ni * np.cos(theta_t)) \
        / (ni * np.cos(theta_t) + nt * np.cos(theta_i))

n = 1_000  # Number of points
theta_i = np.deg2rad(np.linspace(0, 90))
```

```python
# Case A - External reflection
theta_t = np.arcsin(n_air * np.sin(theta_i) / n_water)
theta_B = np.rad2deg(np.arctan(n_water / n_air))

plt.plot(
    np.rad2deg(theta_i),
    r_TE(n_air, theta_i, n_water, theta_t),
    label='$r_{TE}$'
)
plt.plot(
    np.rad2deg(theta_i),
    r_TM(n_air, theta_i, n_water, theta_t),
    label='$r_{TM}$'
)
plt.axvline(
    x=theta_B,
    ls='--', c='k', alpha=0.5,
    label=r'$\theta_B$' + f' = {theta_B:.2f}' + r'$^\circ$'
)
plt.xlabel(r'$\theta_i$ [deg]')
plt.ylabel('Amplitude reflection coefficient')
plt.title('External reflection')
plt.xlim(0, 90)
plt.legend()
plt.show()
```

```python
# Case B - Internal reflection
theta_t = np.arcsin(n_water * np.sin(theta_i) / n_air)
theta_B = np.rad2deg(np.arctan(n_air / n_water))
theta_C = np.rad2deg(np.arcsin(n_air / n_water))

plt.plot(
    np.rad2deg(theta_i),
    r_TE(n_water, theta_i, n_air, theta_t),
    label='$r_{TE}$'
)
plt.plot(
    np.rad2deg(theta_i),
    r_TM(n_water, theta_i, n_air, theta_t),
    label='$r_{TM}$'
)
plt.axvline(
    x=theta_B,
    ls='--', c='k', alpha=0.5,
    label=r'$\theta_B$' + f' = {theta_B:.2f}' + r'$^\circ$'
)
plt.axvline(
    x=theta_C,
    ls='--', c='g', alpha=0.5,
    label=r'$\theta_C$' + f' = {theta_C:.2f}' + r'$^\circ$'
)
plt.xlabel(r'$\theta_i$ [deg]')
plt.ylabel('Amplitude reflection coefficient')
plt.title('Internal reflection')
plt.xlim(0, 90)
plt.legend()
plt.show()
```
